## Background and Prior Work

#### Background

Food safety is always a major concern for public health. According to the North Carolina Government, restaurants are inspected one to four times a year, depending on their menu items and food processes"(1). To conduct inspections, Environmental Health Specialists would visit each restaurant and assess the environment based on the Food Safety Code. The inspection score is based on a 100 points grading system. Each violation is associated with a point value. The restaurant loses points if it violates the food safety code. Based on the score, the restaurant would be assigned to a letter grade ("A": 90-100, "B":80-89, "C": less than 79) (1). After the inspection, the restaurant is required to show its health inspection grade to its customer. Thus, for this project, I am interested in the question that how does a restaurant's open date impact its health inspection score and whether we could predict the health inspection score based on the year of establishment to maximize the inspection efficiency by reducing the number of inspection times of business that meets the requirements.

# Data Cleaning

I started the project with data cleaning. Luckily, the dataset I got is relatively clean. I started with removing all the columns which contains information that is unrelated to my project. I renamed columns so that they are clear to read.

```python
inspections_df = pd.read_csv('data/inspections.csv')
inspections_df.shape
```

```python
inspections_df = inspections_df[['hsisid','date','name','city','state','restaurantopendate',
                                 'days_from_open_date','score']]
inspections_df = inspections_df.rename(columns = {'restaurantopendate':'open_date',
                                                 'days_from_open_date':'open_days'})
inspections_df.head()
```

### Remove Timestamp in Date and Open_date
Since this project does not care about the exact time on the inspection date/open date, I decided to remove all of the timestamp in the data.

```python
def remove_timestamp(str_in):
    if 'T00:00:00Z' in str_in:
        date = str_in.replace('T00:00:00Z',' ')
    else:
        return str_in
    return date
```

```python
inspections_df['date'] = inspections_df['date'].apply(remove_timestamp)
inspections_df['open_date'] = inspections_df['open_date'].apply(remove_timestamp)
inspections_df = inspections_df.dropna(subset = ['open_date','open_days','score'])
inspections_df.head()
```

### Check for Null Values and Drop Negative Open_days

```python
print('Missing Value in open date:' + str(inspections_df['open_date'].isnull().sum()))
print('Missing Value in open days:' + str(inspections_df['open_days'].isnull().sum()))
print('Missing Value in score:' + str(inspections_df['score'].isnull().sum()))
inspections_df = inspections_df[(inspections_df['open_days']>0)]
```

### Extract The Year from Inspection Date and Open_date

```python
def remove_date(str_in):
    if '-' in str_in:
        year = str_in[:4]
    else:
        return str_in
    return year
```

```python
inspections_df = inspections_df.assign(year = inspections_df['date'].apply(remove_date))
inspections_df['open_date'] = inspections_df['open_date'].apply(remove_date)
inspections_df = inspections_df.rename(columns = {'open_date':'open_year'})
inspections_df = inspections_df[['hsisid', 'year','date','name','city','state','open_year','open_days','score']]
inspections_df.head()
```

### Normality Test
Since both two groups of data looks skewed, I did normality tests to check the distribution of the data.

```python
score_test = inspections_df['score'].values
opendays_test = inspections_df['open_days'].values
```

```python
xs = np.arange(score_test.min(), score_test.max(), 0.1)
fit = stats.norm.pdf(xs, np.mean(score_test), np.std(score_test))
plt.plot(xs, fit, label='Normal Dist.', lw=3)
plt.hist(score_test, 50, normed=True, label='Actual Data');
plt.legend();
```

```python
xs = np.arange(opendays_test.min(), opendays_test.max(), 0.1)
fit = stats.norm.pdf(xs, np.mean(opendays_test), np.std(opendays_test))
plt.plot(xs, fit, label='Normal Dist.', lw=3)
plt.hist(opendays_test, 50, normed=True, label='Actual Data');
plt.legend();
```

According to normality tests, the inspection score data is negatively skewed and the days from opening data is positively skewed. Thus, I can not use analysis method that assume normal distribution for these two groups of data.

### Group Data by Years to Analyze Patterns for Each Year From 2012 to 2016

Since our dataset includes inspection scores from the year 2012 to 2016, I am interested in finding out whether the relationship between inspection scores and open days would change over the years. If our hypothesis is true, we would expect the inspection score would be lower in the recent year because restaurants have opened for a longer time compared to the previous year. I used the scatter plot and bar chart to visualize the difference in score and open days across years.

```python
year_fig1 = sns.scatterplot(x ='open_days', y ='score', hue = 'year', alpha=0.4, data=inspections_df, s=50)
year_fig1.set_title('Open Days and Inspections Score for Restaurants in North Carolina (2012-2016)')
year_fig1.set_xlabel('Days from Opening')
year_fig1.set_ylabel('Inspection Score')
```

Based on the scatter plot above, we can see that most restaurants received score ranging in 90-100 for most years. Every year there are some restaurants got inspection scores that are lower than 90. The pattern in each year is similar to the patterns in other years. Since there is a lot of overlapping between dots, I created a bar plot to compare the mean of open days and mean of inspection scores to compare the data across the years.

```python
year_df = inspections_df.groupby(['year']).mean()
year_df = year_df.drop(columns = ['hsisid','open_days'])
year_fig2 = sns.barplot(year_df.index, y='score',data=year_df)
year_fig2.set_title('Average Insepction Score Across the Years (2012-2016)', fontsize=14)
year_fig2.set_xlabel('Year')
year_fig2.set_ylabel('Inspection Score')
```

In this bar plot, we can see that the mean of inspection scores remains same across the years. This shows that even through the open days increases as the year passes, the inspections scores remains similar. This may implies that restaurants are working on maintaining the same high quality of health inpections every year.
