```python
inspections_df = pd.read_csv('inspections.csv')
inspections_df.head()
```

```python
## DEAL WITH INPECTION DATASET


## inspection_df
print("====================get useful attribute====================\n")
print(' before inspection_df column is:',inspections_df.columns)
inspections_df = inspections_df[['hsisid','address1','name','phonenumber','x','y','city','state','zip','score','inspector_id']]
print(' after inspection_df column is:',inspections_df.columns)


print("====================delete null value====================\n")
print("before shape is:", inspections_df.shape)
print('check null value\n', inspections_df.isnull().sum())
inspections_df = inspections_df.dropna()
print('after shape is:', inspections_df.shape)

print("====================round float number====================\n")
print("before:\n",inspections_df[['x','y']].head())
inspections_df[['x','y']] = inspections_df[['x','y']].round(4)
print("after:\n",inspections_df[['x','y']].head())

print("====================no big bias among inspectors====================\n")
print("max mean value per inspectors is", inspections_df.groupby('inspector_id')['score'].mean().max())
print("min mean value per inspectors is", inspections_df.groupby('inspector_id')['score'].mean().min())

print("====================convert phone number format====================\n")
inspections_df = inspections_df[['phonenumber','score']]
print("before\n",inspections_df['phonenumber'].unique())
def convertP(s1):
    s1 = s1[1:4] + s1[6:9] + s1[10:14]
    return s1
inspections_df['phonenumber'] = inspections_df['phonenumber'].apply(convertP)
inspections_df.rename(columns={'phonenumber':'phone'}, inplace=True)
inspections_df = inspections_df.drop_duplicates(subset=['phone'])
print("after\n",inspections_df['phone'].unique())
```

```python
yelp_df = pd.read_csv('yelp.csv')
yelp_df.head()
```

```python
print("====================get useful attribute====================\n")
print(' before yelp_df column is:',yelp_df.columns)
yelp_df = yelp_df.drop(['latitude','longitude','hotdogs','sandwiches','pizza','burgers','burgers','coffee','chicken_wings','delis','bakeries'],axis=1)
print(' after yelp_df column is:',yelp_df.columns)

print("====================check columns====================\n")
print("only care opening restaurant and change price to price level")
yelp_df = yelp_df.loc[yelp_df['is_closed'] == False]
yelp_df = yelp_df.loc[yelp_df['price'].isin(['$','$$','$$$','$$$$'])]
def convert_price_to_level(label):
    if label == '$':
        return 1
    if label == '$$':
        return 2
    if label == '$$$':
        return 3
    if label == '$$$$':
        return 4
yelp_df['price'] = yelp_df['price'].apply(convert_price_to_level)

print("====================delete null value====================\n")
print("before shape is:", yelp_df.shape)
print('check null', yelp_df.isnull().sum())
yelp_df = yelp_df.dropna()
print('after shape is:', yelp_df.shape)

print("====================convert phone number format====================\n")
print("before",yelp_df['phone'].unique())
yelp_df['phone'] = yelp_df['phone'].str[1:]
yelp_df['phone'] = yelp_df['phone'].astype(str)
print('after',yelp_df['phone'].unique())
```

```python
print("====================combine two datasets together====================\n")
df = yelp_df.merge(inspections_df,how='left')
df.head()
```

```python
print("====================delete security value====================\n")
df = df.drop('phone',axis=1)

print("====================delete null value====================\n")
print("before shape is:", df.shape)
print('check null\n', df.isnull().sum())
df = df.dropna()
print('after shape is:', df.shape)
```

# Data Analysis & Results

Before checking the ratings and price variables influence on inspection score, we first do the t-test on the traditional American restaurant and new American restaurant to see is there some correlation on the varaible we will use later. As the result show below, the tradtion American and new American restuarant will influence on the Yelp ratings and Price. However, there are not obviously significant difference on the score, which is dependant vairable we will test later. So we do not find to put them as the new variable to consider. 

```python
rating_traA = df[df['tradamerican'] == True]['rating'].values
rating_newA = df[df['newamerican'] == True]['rating'].values
print("tradamerican mean value is",rating_traA.mean())
print("newamerican mean value is",rating_newA.mean())
t_val ,p_val = ttest_ind(rating_traA,rating_newA)

print("t_val is:",t_val,"\np_val is:", p_val)
# Check if statistical test passes significance, using an alpha value of 0.01. This code provided.
if p_val < 0.01:
    print('Data Science accomplished, there is a significant difference!')
else:
    print('There is NOT a significant difference!')
```

```python
price_traA = df[df['tradamerican'] == True]['price'].values
price_newA = df[df['newamerican'] == True]['price'].values
print("tradamerican mean value is",price_traA.mean())
print("newamerican mean value is",price_newA.mean())
t_val ,p_val = ttest_ind(price_traA,price_newA)

print("t_val is:",t_val,"\np_val is:", p_val)
# Check if statistical test passes significance, using an alpha value of 0.01. This code provided.
if p_val < 0.01:
    print('Data Science accomplished, there is a significant difference!')
else:
    print('There is NOT a significant difference!')
```

```python

score_traA = df[df['tradamerican'] == True]['score'].values
score_newA = df[df['newamerican'] == True]['score'].values
print("tradamerican mean value is",score_traA.mean())
print("newamerican mean value is",score_newA.mean())
t_val ,p_val = ttest_ind(score_traA,score_newA)

print("t_val is:",t_val,"\np_val is:", p_val)
# Check if statistical test passes significance, using an alpha value of 0.01. This code provided.
if p_val < 0.01:
    print('Data Science accomplished, there is a significant difference!')
else:
    print('There is NOT a significant difference!')
```
