```python
#To convert that data into a dataframe that only contains the health inspection id of a restaurant and the number of violations tied to them
vdf = vdf['hsisid'].value_counts().rename_axis('hsisid').reset_index(name='violations')
```

```python
#Merge violations and restaurants dataframes into one then clean into one with just name and violations
rvdf = pd.merge(left=rdf, right=vdf)
rvdf = rvdf[['name', 'violations']]
rvdf
```
